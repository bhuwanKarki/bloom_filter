```python
import os
import json
import binascii
import numpy as np
from bitarray import bitarray
import requests
import time
import statistics
import math
```

```python
#  different number of elements and hashes
N = 1000050000
largePrime = 1113162576
B = bitarray(N)
B.setall(0)
r1=np.random.randint(N)
r2=np.random.randint(N)
r3=np.random.randint(N)
r4=np.random.randint(N)
r5=np.random.randint(N)
r6=np.random.randint(N)
r7=np.random.randint(N)
```

```python
ID=[]
r = requests.get('http://stream.meetup.com/2/rsvps', stream=True)
starttime=time.time()
for raw_rsvp in r.iter_lines():
    if(time.time()-starttime>18000):
        break
    if raw_rsvp:
        rsvp = json.loads(raw_rsvp)
        member_ids=rsvp[u'member'][u'member_id']    
        member_names=rsvp[u'member'][u'member_name']
        hash_member1=((member_ids*r1+r2)%largePrime)%N
        B[hash_member1]=1
        hash_member2=((member_ids*r3+r4)%largePrime)%N
        B[hash_member2]=1
        hash_member3=((member_ids*r5+r6)%largePrime)%N
        B[hash_member3]=1
        print("hashing : {}".format(member_ids))
starttime1=time.time()        
r = requests.get('http://stream.meetup.com/2/rsvps', stream=True)
for raw_rsvp in r.iter_lines():
    if(time.time()-starttime1>18000):
        break
    if raw_rsvp:
    
        rsvp = json.loads(raw_rsvp)
        member_ids=rsvp[u'member'][u'member_id']    
        member_names=rsvp[u'member'][u'member_name']
   
        hash_member1=((member_ids*r1+r2)%largePrime)%N
        hash_member2=((member_ids*r3+r4)%largePrime)%N
        hash_member3=((member_ids*r5+r6)%largePrime)%N
        print("verify : {}".format(member_ids))
        if(B[hash_member1]==1) and (B[hash_member2]==1) and (B[hash_member3]==1):
            ID.append([hash_member1,hash_member2,hash_member3])
            print("{} logged one day ago".format(member_ids))
        

```
